# Grafo de Hora de Aventura 

Hora de Aventura es una serie animada de televisión creada por Pendleton Ward en el 2010 para Cartoon Network. La serie sigue las aventuras de Finn, un niño, y Jake, un perro con poderes mágicos con los que puede cambiar de forma, crecer y encoger a voluntad, que habitan en la post-apocalíptica Tierra de Ooo.

La serie es conocida por su estilo surrealista, su humor peculiar y sus personajes únicos. Aunque muchas aventuras son autoconclusivas, hay tramas continuas y desarrollo de personajes a lo largo de las temporadas. "Hora de Aventura" explora temas profundos como la amistad, el crecimiento personal, la pérdida y la identidad, presentados de una manera accesible y entretenida para una audiencia tanto joven como adulta.

# Trama
La serie está ambientada en la Tierra de Ooo, un mundo post-apocalíptico lleno de criaturas mágicas, reinos variados, y personajes excéntricos. Cada episodio presenta una nueva aventura o desafío para Finn y Jake, donde se enfrentan a diferentes enemigos, resuelven problemas, y ayudan a sus amigos.

# Personajes Principales
Finn el Humano: Un valiente y entusiasta niño que lucha contra el mal y busca aventuras.

Jake el Perro: El mejor amigo y hermano adoptivo de Finn, con la habilidad de estirarse y cambiar de forma.

Dulce Princesa: La gobernante del Reino Dulce, una científica brillante con una apariencia de chicle.

Rey Helado: Un villano cómico y trágico que usa una corona mágica para controlar el hielo y busca desesperadamente compañía.

Marceline la Reina Vampiro: Una vampira y músico que tiene una historia compleja y profunda con varios personajes.

BMO: Un pequeño robot con inteligencia artificial que actúa como consola de videojuegos, computadora y amigo de Finn y Jake.

 ### En Hora de Aventura, los personajes tienen diversas relaciones entre sí, que pueden ser de amistad, enemistad, o alianzas

Finn y Jake: Finn el Humano y Jake el Perro son los mejores amigos y hermanos adoptivos. Viven juntos y comparten muchas aventuras.

Finn y Dulce Princesa (Princess Bubblegum): Finn está enamorado de Dulce Princesa, y ella lo ve como un gran amigo y protector del Reino Dulce.

Finn y Marceline: Finn tiene una relación de amistad con Marceline la Reina Vampiro. Han compartido varias aventuras y tienen una conexión cercana.

Jake y Dulce Princesa: Jake también es amigo de Dulce Princesa y la ayuda en varias ocasiones.

Jake y BMO: Jake y BMO, el robot consola de videojuegos, viven juntos y tienen una relación de compañeros de cuarto y amigos.

Dulce Princesa y Marceline: Dulce Princesa y Marceline tienen una relación compleja que ha evolucionado desde la enemistad a una profunda amistad.

Marceline y Rey Helado (Ice King): Marceline y Rey Helado tienen una historia compartida que data de mucho antes de los eventos principales de la serie.
Aunque Rey Helado es visto como un villano, Marceline tiene un lugar especial para él en su corazón debido a su pasado común.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Creacion de el grafo

```python

G = nx.Graph()

personajes = [
    'Finn el Humano', 'Jake el Perro', 'Dulce Princesa', 'Rey Helado', 'Marceline', 'BMO',
    'Reina Grumosa', 'Rey de Ooo', 'Chicle Guardián', 'Princesa Flama', 'Lady Rainicorn', 
    'Gunther', 'Prismo', 'Ricardio el Corazón'
]

for personaje in personajes:
    G.add_node(personaje)

# Añadir relaciones
relaciones = [
    ('Finn el Humano', 'Jake el Perro'),
    ('Finn el Humano', 'Dulce Princesa'),
    ('Finn el Humano', 'Marceline'),
    ('Finn el Humano', 'Princesa Flama'),
    ('Jake el Perro', 'Dulce Princesa'),
    ('Jake el Perro', 'BMO'),
    ('Jake el Perro', 'Lady Rainicorn'),
    ('Jake el Perro', 'Prismo'),
    ('Dulce Princesa', 'Marceline'),
    ('Dulce Princesa', 'Chicle Guardián'),
    ('Rey Helado', 'Marceline'),
    ('Rey Helado', 'Gunther'),
    ('Reina Grumosa', 'Finn el Humano'),
    ('Reina Grumosa', 'Jake el Perro'),
    ('Rey de Ooo', 'Dulce Princesa'),
    ('Ricardio el Corazón', 'Finn el Humano'),
    ('Ricardio el Corazón', 'Dulce Princesa')
]

for relacion in relaciones:
    G.add_edge(*relacion)

```

```python

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=2000)


nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='gray')

nx.draw_networkx_labels(G, pos, font_size=12, font_family='sans-serif')

plt.title('Grafo de Personajes de Hora de Aventura')
plt.show()
```
