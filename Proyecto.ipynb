# Grafo de Hora de Aventura 

Hora de Aventura es una serie animada de televisión creada por Pendleton Ward en el 2010 para Cartoon Network. La serie sigue las aventuras de Finn, un niño, y Jake, un perro con poderes mágicos con los que puede cambiar de forma, crecer y encoger a voluntad, que habitan en la post-apocalíptica Tierra de Ooo.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Crear un grafo vacío
G = nx.Graph()

# Añadir nodos (personajes)
G.add_node('Finn')
G.add_node('Jake')
G.add_node('Princess Bubblegum')
G.add_node('Ice King')
G.add_node('Marceline')
G.add_node('BMO')

# Añadir aristas (relaciones)
G.add_edge('Finn', 'Jake')
G.add_edge('Finn', 'Princess Bubblegum')
G.add_edge('Finn', 'Marceline')
G.add_edge('Jake', 'Princess Bubblegum')
G.add_edge('Jake', 'BMO')
G.add_edge('Princess Bubblegum', 'Marceline')
G.add_edge('Ice King', 'Marceline')
```

```python
# Configurar el diseño del grafo
pos = nx.spring_layout(G)

# Dibujar nodos
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=2000)

# Dibujar aristas
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='gray')

# Dibujar etiquetas
nx.draw_networkx_labels(G, pos, font_size=12, font_family='sans-serif')

# Mostrar el grafo
plt.title('Grafo de Personajes de Hora de Aventura')
plt.show()
```

```python
# Calcular la centralidad de grado (número de conexiones de cada nodo)
degree_centrality = nx.degree_centrality(G)
print("Centralidad de Grado de los Nodos:")
for node, centrality in degree_centrality.items():
    print(f"{node}: {centrality:.2f}")

# Encontrar componentes conexas
components = list(nx.connected_components(G))
print("\nComponentes Conexas:")
for i, component in enumerate(components):
    print(f"Componente {i + 1}: {component}")
```
